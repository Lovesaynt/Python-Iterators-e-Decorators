# DECORATORS

Decorators são funções que são chamadas antes de outras funções para trabalhar em cima delas.

Ex: 
    Função A
    
    @Função A
    Função B

Função A está sendo aplicada em cima da Função B

```python
def meu_decorator(func):
    def wrapper():
        print("Antes da função")
        func()
        print("Depois da função")
    return wrapper


@meu_decorator
def diga_oi():
    print("Oi!")

diga_oi()
```
