# DECORATORS

Decorators são funções que são chamadas antes de outras funções para trabalhar em cima delas.

Ex: 
    Função A
    
    @Função A
    Função B

Função A está sendo aplicada em cima da Função B

```python
def meu_decorator(func): # 
    def wrapper1(): # Definindo nosso Wrapper.
        print("Antes da função")
        func() # Chamandoa função onde nosso decorator irá trabalhar.
        print("Depois da função")
    return wrapper1 # Retornando nosso Wrapper.


@meu_decorator # Função A
def diga_oi(): # Função B
    print("Oi!")

diga_oi()
```

# Importante!

- O nome Wrapper é opcional, porém esse é o mais comum.
- No final do seu decorator retorne sempre seu Wrapper
- Não se esqueça de chamar a função filho dentro do Wrapper.

    Outras maneiras de usar Decorators

```python
def meu_decorator2(func):
    def wrapper2(*args, **kwargs):
        print('Código Antes')
        resultado = func(*args, **kwargs)
        print('Código Depois')
        return resultado
    return wrapper2

def saudacao(nome):
    print(f"Olá, {nome}!")

saudacao = meu_decorator2(saudacao) # Substitui a necessidade do @Decorator

saudacao(nome='Pablo')
```

### Decorator com Parâmetro

```python
def repetir(n):
    def decorator(func): # Nosso Decorator
        def wrapper3(*args, **kwargs): # Nosso Wrapper
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper3
    return decorator

@repetir(3) # Passando o parâmetro de repetições do nosso Decorator
def oi():
    print("Oi")

oi()
```
