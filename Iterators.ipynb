# ITERATORS

    Iterators em Python são objetos que permitem percorrer uma sequência de elementos um por vez.

```python
lista = [1, 2, 3]

iterador = iter(lista) # cria um iterator

print(next(iterador)) # 1
print(next(iterador)) # 2
print(next(iterador)) # 3
print(next(iterador)) # Exceção StopIteration
```

### Como funciona no for:

```python
for item in [1, 2, 3]:
    print(item)
```

    Equivale a:

```python
it = iter([1, 2, 3])
while True:
    try:
        item = next(it)
        print(item)
    except StopIteration:
        break
```
